```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
x= np.linspace(0,5,11)
x
y= x**2
y
```

```python
plt.plot(x,y)
plt.title("hello")
plt.xlabel("x")
plt.ylabel("y")
```

subplot in matplot

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

object orianted way

```python
fig= plt.figure(figsize=(12,8),dpi=100)
axes1= fig.add_axes([0.8 , 0.8 , 1, 1])

axes1.plot(x,y)
axes2= fig.add_axes([0.8 , 0.8 , .8, .7])
axes2.plot(x,y)
```

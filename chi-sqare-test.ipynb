```python
import numpy as np 
import pandas as pd 
# from scipy.stats import chisquare
import seaborn as sns 

```

```python
observed = [8, 12, 9, 11, 10, 10]
expected = [10, 10, 10, 10, 10, 10]
```

```python
chi2_test =0
for o , e  in zip(observed, expected):
    chi2_test += (o - e)**2 / e

```

```python
print(f"chi-sqare-test is {chi2_test}")
```

```python
df = len(observed)-1
print(f'df is {df}')
```

```python
table_value = 11.07


if chi2_test < table_value:
    print("Result: Accept H0 (Dice is Fair)")
else:
    print("Result: Reject H0 (Dice is Not Fair)")
```

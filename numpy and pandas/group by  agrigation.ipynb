```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    "Date": pd.date_range(start="2024-01-01", periods=10, freq="D"),
    "Category": ["Food", "Travel", "Food", "Shopping", "Travel", 
                 "Food", "Shopping", "Food", "Travel", "Shopping"],
    "Amount": [200, 500, 150, 800, 300, 250, 700, 100, 400, 600]
})
```

```python
df
```

```python
cat =df.groupby("Category")['Amount'].sum()

cat
```

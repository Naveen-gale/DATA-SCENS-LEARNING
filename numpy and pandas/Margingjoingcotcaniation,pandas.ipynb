```python
import numpy as np
import pandas as pd
```

```python
students = pd.DataFrame({
    "Student_ID": [1, 2, 3, 4, 5],
    "Name": ["Naveen", "Rahul", "Priya", "Anjali", "Kiran"],
    "Age": [20, 21, 19, 22, 20]
})

marks = pd.DataFrame({
    "Student_ID": [1, 2, 3, 6],
    "Marks": [85, 78, 92, 88],
    "Grade": ["A", "B", "A", "B"]
})
students
```

```python
pd.merge(students,marks)
```

```python
pd.merge(students ,marks, on="Student_ID" ,how ="outer")   #mearg all 
```

```python
pd.merge(students ,marks, on="Student_ID" ,how ="left")   #lefte = student
```

```python
pd.merge(students ,marks, on="Student_ID" ,how ="right")  #right = marks
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Naveen", "Rahul", "Priya"],
    "Marks": [85, 78, 92]
})
df2 = pd.DataFrame( {
    "ID": [4, 5, 6],
    "Name": ["Anjali", "Kiran", "Arun"],
    "Marks": [74, 88, 81]
})
```

```python
df1
```

```python
pd.concat([df1,df2]) #based on columne
```

```python
pd.concat([df1,df2],axis=1)
```

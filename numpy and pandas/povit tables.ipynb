** povit tables**

```python
import numpy as np
import pandas as pd
```

```python

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Subject": ["Math", "Science", "English"],
    "Marks": [90, 80, 85]
})
df
```

```python
pivot = df.pivot(index="Name", columns="Subject", values="Marks")
print(pivot)
```
